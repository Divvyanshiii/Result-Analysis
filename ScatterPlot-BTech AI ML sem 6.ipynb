```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'B.Tech Semester 6 Publish Result File.xlsx'  # replace with your actual file path
df = pd.read_excel(file_path, sheet_name='result (14)')  # adjust sheet name if needed 

# Display the first few rows of the dataframe
df.head(10)

```

```python
# Dropping irrelevant ROWS i.e., Subject Name, Credit Point, Student roll number from above
df_cleaned = df.drop([0,1,2]).reset_index(drop=True)
df_cleaned
```

```python
df_cleaned.columns
```

## Data Preprocessing 

```python
new_column_names = {
#   'Column name in above df' : 'name we want to rename it to'
    
#     'Sub Code': 'Compiler Design',  # Example: '15BCS-0CD32T' renamed to 'Compiler Design'
    'Sub Code': 'Roll Number',
    'Unnamed: 1': 'Student Name',
    'Unnamed: 2': 'SGPA',
    'Unnamed: 3': 'Exam Name',
    '15BCS-0CD32T': 'Compiler Design M', # M - Marks
    'Unnamed: 5': 'Compiler Design G', # G - Grades
    '15BCS-0MP41M': 'Minor Project M',
    'Unnamed: 7': 'Minor Project G',
    '15BCS-0TT42C': 'Technology Trends-4 M',
    'Unnamed: 9': 'Technology Trends-4 G',
    '15BCS-1CN31T': 'Computer Networks M',
    'Unnamed: 11': 'Computer Networks G',
    '18BCS-0AI42E': 'AI KRR M',
    'Unnamed: 13': 'AI KRR G',
    '21BTC-1SS31C': 'Service Learning M',
    'Unnamed: 15': 'Service Learning G',
    '22BCS-0MC32E': 'Machine Learning-2 M',
    'Unnamed: 17': 'Machine Learning-2 G',
    '24BCS-0CP32E': 'Competitive Programming M',
    'Unnamed: 19': 'Competitive Programming G',
    'TDL': 'TDL M',
    'Unnamed: 21': 'TDL G'
#     'TDL Code' : 
}

# Rename columns using the dictionary
df_cleaned = df_cleaned.rename(columns=new_column_names)
df_cleaned
```

```python
df_cleaned.drop(columns=['TDL Code'])
# We do not need this column for our analysis 
```

```python
# Separate internal and external marks
# in our dataframe 'df_cleaned' , under column 'Exam Name' if value is 'Internal' and 'External'
df_internal = df_cleaned[df_cleaned['Exam Name'] == 'Internal'].reset_index(drop=True)
df_external = df_cleaned[df_cleaned['Exam Name'] == 'External'].reset_index(drop=True)

# Initialize new dataframe 

# This DF will be used for making the process of plotting the scatterplot easier
columns = ['Roll Number', 'Student Name']
subjects = ['Compiler Design', 'Minor Project', 'Technology Trends-4', 
            'Computer Networks', 'AI KRR', 'Service Learning', 
            'Machine Learning-2', 'Competitive Programming','TDL']

df_new = pd.DataFrame(columns=columns)

# Fill the new dataframe
for index, row_internal in df_internal.iterrows():
    new_row = {'Roll Number': row_internal['Roll Number'], 'Student Name': row_internal['Student Name']}
    for subject in subjects:
        internal_marks = row_internal[f'{subject} M']
        external_marks = df_external.loc[index, f'{subject} M']
        new_row[f'{subject}_Internal'] = internal_marks
        new_row[f'{subject}_External'] = external_marks
    df_new = pd.concat([df_new, pd.DataFrame([new_row])], ignore_index=True)
    
df_new = df_new.drop(columns=['Roll Number','Student Name'])
df_new
```

### ScatterPlot

```python
#  using matplotlib

for subject in subjects:
    internal_col = f'{subject}_Internal'
    external_col = f'{subject}_External'
    
    # Create scatter plot
    plt.figure()
    plt.scatter(df_new[internal_col], df_new[external_col])
    
    plt.title(f'Scatterplot of {subject}')
    plt.xlabel('Internal Marks')
    plt.ylabel('External Marks')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f'{subject}_scatterplot.png')
    plt.show()
```

```python
# using seaborn

import seaborn as sns
for subject in subjects:
    internal_col = f'{subject}_Internal'
    external_col = f'{subject}_External'
    
    # Create scatter plot using Seaborn
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=internal_col, y=external_col, data=df_new)
    plt.title(f'Scatterplot of {subject}')
    plt.xlabel('Internal Marks')
    plt.ylabel('External Marks')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
