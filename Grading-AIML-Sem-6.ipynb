```python
grade_order = ['C', 'D', 'F']

def highlight_failing_rows(row):
    if 'F' in row.values:
        return ['background-color: red'] * len(row)
    elif 'D' in row.values:
        return ['background-color: orange'] * len(row)
    else:
        return ['background-color: yellow'] * len(row)
    
    

# Iterate over each subject
subjects = df.columns[2:]  # Assuming grades start from column index 2

for subject in subjects:
    # Filter students with grades C, D, or F in the current subject and sort by grade_order
    filtered_students = df[df[subject].isin(grade_order)].copy()
    filtered_students['Grade'] = pd.Categorical(filtered_students[subject], categories=grade_order, ordered=True)
    filtered_students.sort_values(by='Grade', inplace=True)
    
    # Create a new DataFrame for the current subject
    results_df = filtered_students[['Roll Number', 'Student Name', subject]].copy()
    results_df.columns = ['Roll Number', 'Student Name', 'Grade']

    # Display the results for the current subject using display()
    print(f"Subject: {subject}")
#     display(results_df)
    display(results_df.style.apply(highlight_failing_rows, axis=1))
    print()
```

```python
import openpyxl
from openpyxl.styles import PatternFill

grade_order = ['C', 'D', 'F']

# Custom function to highlight grades
def highlight_grades(row):
    if 'F' in row.values:
        return ['background-color: red'] * len(row)
    elif 'D' in row.values:
        return ['background-color: orange'] * len(row)
    elif 'C' in row.values:
        return ['background-color: yellow'] * len(row)
    else:
        return [''] * len(row)

# Iterate over each subject
subjects = df.columns[2:]  # Assuming grades start from column index 2

# Create a master DataFrame to store all results
master_df = pd.DataFrame(columns=['Roll Number', 'Student Name', 'Grade', 'Subject'])

for subject in subjects:
    # Filter students with grades C, D, or F in the current subject and sort by grade_order
    filtered_students = df[df[subject].isin(grade_order)].copy()
    filtered_students['Grade'] = pd.Categorical(filtered_students[subject], categories=grade_order, ordered=True)
    filtered_students.sort_values(by='Grade', inplace=True)
    
    # Create a new DataFrame for the current subject
    results_df = filtered_students[['Roll Number', 'Student Name', subject]].copy()
    results_df.columns = ['Roll Number', 'Student Name', 'Grade']

    # Add subject column to the results DataFrame
    results_df['Subject'] = subject

    # Append to master DataFrame
    master_df = pd.concat([master_df, results_df], ignore_index=True)

# Save the master DataFrame to an Excel file
output_file = 'subject_wise_grades.xlsx'
master_df.to_excel(output_file, index=False)

# Load the workbook and the sheet to apply color coding
wb = openpyxl.load_workbook(output_file)
ws = wb.active

# Define fill colors for grades
fills = {
    'F': PatternFill(start_color='FF0000', end_color='FF0000', fill_type='solid'),
    'D': PatternFill(start_color='FFA500', end_color='FFA500', fill_type='solid'),
    'C': PatternFill(start_color='FFFF00', end_color='FFFF00', fill_type='solid')
}

# Apply color coding
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=ws.max_column):
    grade = row[2].value  # Grade is in the third column
    if grade in fills:
        for cell in row:
            cell.fill = fills[grade]

# Save the workbook
wb.save(output_file)

# Display the final DataFrame with color coding (for Jupyter Notebooks)
def display_color_coded_df(df):
    def color_row(row):
        color = ''
        if row['Grade'] == 'F':
            color = 'background-color: red'
        elif row['Grade'] == 'D':
            color = 'background-color: orange'
        elif row['Grade'] == 'C':
            color = 'background-color: yellow'
        return [color] * len(row)
    
    return df.style.apply(color_row, axis=1)

display(display_color_coded_df(master_df))
```
